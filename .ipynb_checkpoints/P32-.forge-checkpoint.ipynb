# forge 2クラス分類データセット

```python
import mglearn
import mglearn.datasets as mgdata # cachedir='cache' を location='cache'に変更
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# forgeデータセット
X, y = mgdata.make_forge()
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['Class 0', 'Class 1'], loc=4)
plt.xlabel('First feature')
plt.ylabel('Second feature')
print('X.shape: {}'.format(X.shape))
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1) # 近傍点1
```

1-最近傍モデルによるforgeデータセットに対する予測

```python
mglearn.plots.plot_knn_classification(n_neighbors=3) # 近傍点3
```

3-最近傍モデルによるforgeデータセットに対する予測

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # データセット分割
clf = KNeighborsClassifier(n_neighbors=3) # インスタンスオブジェクトの作成
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

約86% の精度

# KNeighborsClassifier 決定境界線の比較

```python
# KNeighborsClassifier の解析
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y) # 近傍点1, 3, 9 点のモデル作成
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4) # 第一引数:モデル 第二引数:データ
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax) # 第一引数:X値 第二引数:Y値 第三引数:データラベル 第四引数:タイトル,X軸,Y軸ラベル
    ax.set_title('{} neighbor(s)'.format(n_neighbors))
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')

axes[0].legend(loc=3)
```

決定境界が描画されている  
最近傍点が少ない場合 複雑度の高いモデル  
最近傍点が多い場合 複雑度の低いモデル
