```python
import numpy as np
import mglearn
import mglearn.datasets as mgdata # cachedir='cache' を location='cache'に変更
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor # 回帰モデル
```

```python
# waveデータセット
X, y = mgdata.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel('Feature')
plt.ylabel('Target')
```

```python
print(y[:5])
```

# 最近傍回帰

```python
mglearn.plots.plot_knn_regression(n_neighbors=4) # 最近傍点数を変更して確認
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print(reg.predict(X_test))
print(reg.score(X_test, y_test))
```

83% の精度

# KNeighborsRegressor の解析

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
line = np.linspace(-3, 3, 1000).reshape(-1, 1) # -3 ~ 3 の間に1000点のポイントを作る
for n_neighbors, ax in zip([1, 3, 9], axes): # 近傍点数1, 3, 9
    reg = KNeighborsRegressor(n_neighbors=n_neighbors) 
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title('{} neighbor(s)\n train score: {:.2f} test score: {:.2f}'.format(n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel('Feature')
    ax.set_ylabel('Trget')

axes[0].legend(['Model predictions', 'Training data/target', 'Test data/target'], loc='best')
```

近傍点が1点の場合は予測はすべての訓練データを通る事になる為、複雑であり不安定  
近傍点が増えていくとシンプルになり予想はスムーズになるが、訓練データの精度は下がる
